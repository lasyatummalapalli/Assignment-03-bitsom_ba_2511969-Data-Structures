```python
# ----- Given menu (same as above) -----
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# cart as list of dicts
cart = []

def find_cart_index(item_name):
    """Return index of item in cart, or -1 if not present."""
    for i in range(len(cart)):
        if cart[i]["item"] == item_name:
            return i
    return -1

def add_item_to_cart(item_name, qty):
    if item_name not in menu:
        print(f"Error: '{item_name}' does not exist in menu.")
        return

    item_info = menu[item_name]

    if not item_info["available"]:
        print(f"Error: '{item_name}' is marked unavailable.")
        return

    idx = find_cart_index(item_name)
    if idx != -1:
        cart[idx]["quantity"] += qty
        print(f"Increased '{item_name}' quantity to {cart[idx]['quantity']}.")
    else:
        cart.append({
            "item": item_name,
            "quantity": qty,
            "price": item_info["price"]
        })
        print(f"Added new item '{item_name}' x{qty} to cart.")

def remove_item_from_cart(item_name):
    idx = find_cart_index(item_name)
    if idx == -1:
        print(f"Cannot remove: '{item_name}' not in cart.")
    else:
        removed = cart.pop(idx)
        print(f"Removed '{removed['item']}' from cart.")

def change_item_quantity(item_name, qty):
    idx = find_cart_index(item_name)
    if idx == -1:
        print(f"Cannot update: '{item_name}' not in cart.")
        return
    cart[idx]["quantity"] = qty
    print(f"Set '{item_name}' quantity to {qty}.")

def display_cart():
    print("\nCurrent Cart:")
    if not cart:
        print("(Cart is empty)")
        return
    for entry in cart:
        line_total = entry["quantity"] * entry["price"]
        print(f"{entry['item']:<18} x{entry['quantity']}   ₹{line_total:.2f}")

# --------- Simulated operations ----------
add_item_to_cart("Paneer Tikka", 2)
display_cart()

add_item_to_cart("Gulab Jamun", 1)
display_cart()

add_item_to_cart("Paneer Tikka", 1)   # should update to 3
display_cart()

add_item_to_cart("Mystery Burger", 1) # not in menu
add_item_to_cart("Chicken Wings", 1)  # unavailable

remove_item_from_cart("Gulab Jamun")
display_cart()

# --------- Final Order Summary ----------
print("\n========== Order Summary ==========")
subtotal = 0.0

for entry in cart:
    item_total = entry["quantity"] * entry["price"]
    subtotal += item_total
    print(f"{entry['item']:<18} x{entry['quantity']}   ₹{item_total:.2f}")

print("------------------------------------")
gst = subtotal * 0.05
total_payable = subtotal + gst

print(f"Subtotal:                ₹{subtotal:.2f}")
print(f"GST (5%):                ₹{gst:.2f}")
print(f"Total Payable:           ₹{total_payable:.2f}")
print("====================================")
```

```python
import copy

# ----- Step 1: Create deep copy of inventory -----
inventory_backup = copy.deepcopy(inventory)

# ----- Step 2: Modify one value to show independence -----
inventory["Paneer Tikka"]["stock"] = 1

print("After changing live inventory stock for Paneer Tikka:")
print("Live inventory :", inventory["Paneer Tikka"])
print("Backup copy    :", inventory_backup["Paneer Tikka"])

# Restore original state using backup
inventory = copy.deepcopy(inventory_backup)

# ----- Step 3: Final cart from Task 2 -----
cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0}
]

# ----- Step 4: Function to apply cart to inventory -----
def apply_cart_to_inventory(cart_list, inv_dict):
    for row in cart_list:
        item_name = row["item"]
        required = row["quantity"]

        if item_name not in inv_dict:
            print(f"Stock update skipped: {item_name} not in inventory.")
            continue

        available = inv_dict[item_name]["stock"]

        if available >= required:
            inv_dict[item_name]["stock"] = available - required
        else:
            print(
                f"Warning: Only {available} unit(s) available for {item_name}; "
                f"needed {required}. Deducting all and setting stock to 0."
            )
            inv_dict[item_name]["stock"] = 0

# Apply deduction
apply_cart_to_inventory(cart, inventory)

# ----- Step 5: Reorder alerts after deduction -----
print("\nReorder Alerts:")
for name, info in inventory.items():
    if info["stock"] <= info["reorder_level"]:
        print(
            f"⚠ Reorder Alert: {name} — "
            f"Only {info['stock']} unit(s) left (reorder level: {info['reorder_level']})"
        )

# ----- Step 6: Print both dictionaries to show difference -----
print("\nFinal Inventory (after deduction):")
print(inventory)

print("\nInventory Backup (original state):")
print(inventory_backup)
```

```python
# ----- Step 1: Revenue per day -----
print("Revenue per day:")

day_totals = {}

for date, orders in sales_log.items():
    day_total = sum(o["total"] for o in orders)
    day_totals[date] = day_total
    print(f"{date} → ₹{day_total:.2f}")

# ----- Step 2: Best‑selling day -----
best_day = max(day_totals, key=day_totals.get)
print(f"\nBest-selling day: {best_day} ₹{day_totals[best_day]:.2f}")

# ----- Step 3: Most ordered item -----
item_freq = {}

for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            if item not in item_freq:
                item_freq[item] = 0
            item_freq[item] += 1

top_item = max(item_freq, key=item_freq.get)
print("Most ordered item:", top_item)

# ----- Step 4: Add new day to sales_log -----
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

# ----- Step 5: Updated revenue and best day -----
print("\nUpdated Revenue per day:")

day_totals_updated = {}

for date, orders in sales_log.items():
    total_value = 0
    for order in orders:
        total_value += order["total"]
    day_totals_updated[date] = total_value
    print(f"{date} → ₹{total_value:.2f}")

best_day_updated = max(day_totals_updated, key=day_totals_updated.get)
print(f"Updated Best-selling day: {best_day_updated} ₹{day_totals_updated[best_day_updated]:.2f}")

# ----- Step 6: Numbered list of all orders using enumerate -----
print("\nAll Orders:")

counter = 1
for date, orders in sales_log.items():
    for order in orders:
        items_joined = ", ".join(order["items"])
        print(f"{counter}. [{date}] Order #{order['order_id']} — ₹{order['total']:.2f} — Items: {items_joined}")
        counter += 1
```
